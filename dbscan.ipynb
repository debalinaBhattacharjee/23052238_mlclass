```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# ----------------------------
# Load Dataset
# ----------------------------
data = pd.read_csv("Mall_Customers.csv")

# Select features
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ----------------------------
# Apply DBSCAN
# ----------------------------
dbscan = DBSCAN(eps=0.5, min_samples=5)

clusters = dbscan.fit_predict(X_scaled)

# Add cluster column
data['Cluster'] = clusters

# ----------------------------
# Save Output Dataset
# ----------------------------
data.to_csv("dbscan_clustered_customers.csv", index=False)

print("Clustered data saved as dbscan_clustered_customers.csv")

# ----------------------------
# Plot Graph
# ----------------------------
plt.figure()

plt.scatter(X['Annual Income (k$)'],
            X['Spending Score (1-100)'],
            c=clusters)

plt.title("DBSCAN Customer Segmentation")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")

# Save graph
plt.savefig("dbscan_cluster_graph.png")

plt.show()

print("Graph saved as dbscan_cluster_graph.png")
```
