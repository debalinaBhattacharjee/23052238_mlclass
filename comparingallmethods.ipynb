```python
# ================================
# Advertising Dataset – ML Models
# Output + Graph Stored
# ================================

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import mean_squared_error, accuracy_score

# -----------------------------
# Load Dataset
# -----------------------------
data = pd.read_csv("Advertising.csv")

X = data[['TV', 'Radio', 'Newspaper']]
y_reg = data['Sales']
y_clf = (data['Sales'] >= data['Sales'].mean()).astype(int)

# -----------------------------
# Train-Test Split
# -----------------------------
X_train, X_test, y_reg_train, y_reg_test = train_test_split(
    X, y_reg, test_size=0.2, random_state=42)

_, _, y_clf_train, y_clf_test = train_test_split(
    X, y_clf, test_size=0.2, random_state=42)

# -----------------------------
# Feature Scaling
# -----------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# -----------------------------
# Open Output File
# -----------------------------
with open("output.txt", "w") as f:

    f.write("MODEL PERFORMANCE RESULTS\n")
    f.write("=========================\n\n")

    # -------- REGRESSION --------
    reg_results = {}

    lr = LinearRegression()
    lr.fit(X_train, y_reg_train)
    reg_results["Linear Regression"] = mean_squared_error(
        y_reg_test, lr.predict(X_test))

    mlr = LinearRegression()
    mlr.fit(X_train, y_reg_train)
    reg_results["Multiple Linear Regression"] = mean_squared_error(
        y_reg_test, mlr.predict(X_test))

    knn_reg = KNeighborsRegressor(n_neighbors=5)
    knn_reg.fit(X_train_scaled, y_reg_train)
    reg_results["KNN Regression"] = mean_squared_error(
        y_reg_test, knn_reg.predict(X_test_scaled))

    dt_reg = DecisionTreeRegressor(random_state=42)
    dt_reg.fit(X_train, y_reg_train)
    reg_results["Decision Tree Regression"] = mean_squared_error(
        y_reg_test, dt_reg.predict(X_test))

    f.write("REGRESSION RESULTS (MSE)\n")
    for model, mse in reg_results.items():
        f.write(f"{model}: {mse:.4f}\n")

    # -------- CLASSIFICATION --------
    f.write("\nCLASSIFICATION RESULTS (Accuracy)\n")

    nb = GaussianNB()
    nb.fit(X_train, y_clf_train)
    f.write(f"Naive Bayes: {accuracy_score(y_clf_test, nb.predict(X_test)):.4f}\n")

    knn_clf = KNeighborsClassifier(n_neighbors=5)
    knn_clf.fit(X_train_scaled, y_clf_train)
    f.write(f"KNN Classification: {accuracy_score(y_clf_test, knn_clf.predict(X_test_scaled)):.4f}\n")

    dt_clf = DecisionTreeClassifier(random_state=42)
    dt_clf.fit(X_train, y_clf_train)
    f.write(f"Decision Tree Classification: {accuracy_score(y_clf_test, dt_clf.predict(X_test)):.4f}\n")

# -----------------------------
# Graph Generation
# -----------------------------
plt.figure(figsize=(10, 5))
plt.bar(reg_results.keys(), reg_results.values())
plt.xticks(rotation=45, ha="right")
plt.ylabel("Mean Squared Error")
plt.title("Regression Model Comparison")
plt.tight_layout()
plt.savefig("model_comparison.png")
plt.close()

```
